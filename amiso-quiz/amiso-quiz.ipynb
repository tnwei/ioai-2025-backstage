```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
Y = iris.target

# Create an instance of Logistic Regression Classifier and fit the data.
logreg = LogisticRegression(C=1e5)
logreg.fit(X, Y)
```

```python
_, ax = plt.subplots(figsize=(4, 3))
DecisionBoundaryDisplay.from_estimator(
    logreg,
    X,
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="Sepal length",
    ylabel="Sepal width",
    eps=0.5,
)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.Paired)

plt.show()
```

```python
_, ax = plt.subplots(figsize=(6, 4))
plt.xlim(-1, 10)
plt.ylim(-1, 6)

DecisionBoundaryDisplay.from_estimator(
    logreg,
    X,
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="Sepal length",
    ylabel="Sepal width",
    eps=5,
)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.Paired)

plt.show()
```

```python
df1 = pd.read_csv("col1.csv")
df1["label"] = 0
```

```python
df2 = pd.read_csv("col2.csv")
df2["label"] = 1
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.tail()
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="rbf", C=10)
svc.fit(df[["X", "Y"]], df["label"])
```

```python
_, ax = plt.subplots(figsize=(8, 6))

DecisionBoundaryDisplay.from_estimator(
    svc,
    df[["X", "Y"]],
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="X",
    ylabel="Y",
    eps=10,
)
plt.scatter(df["X"], df["Y"], c=df["label"], edgecolors="k", cmap=plt.cm.Paired)

plt.show()
```

```python
_, ax = plt.subplots(figsize=(8, 6))

DecisionBoundaryDisplay.from_estimator(
    svc,
    df[["X", "Y"]],
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="X",
    ylabel="Y",
    eps=10,
)

plt.ylim(0, 130)
plt.xlim(0, 150)

plt.scatter(df["X"], df["Y"], c=df["label"], edgecolors="k", cmap=plt.cm.Paired)

plt.savefig("amiso-quiz-raw.png")
plt.show()
```

```python
_, ax = plt.subplots(figsize=(6, 4))

DecisionBoundaryDisplay.from_estimator(
    svc,
    df[["X", "Y"]],
    cmap=plt.cm.Paired,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    xlabel="X",
    ylabel="Y",
    eps=100,
)
plt.scatter(df["X"], df["Y"], c=df["label"], edgecolors="k", cmap=plt.cm.Paired)

plt.show()
```
